```python
from nba_api.stats.endpoints import teamdashboardbygeneralsplits
import time

time.sleep(0.5)

# Get Lakers stats (use Lakers ID from your teams list)
lakers_stats = teamdashboardbygeneralsplits.TeamDashboardByGeneralSplits(
    team_id=1610612747,  # Lakers ID
    season='2024-25'
)

stats = lakers_stats.get_data_frames()[0]
print(stats[['GP', 'FG_PCT', 'FT_PCT', 'FG3_PCT', 'AST', 'REB']])

# Calculate per game
gp = stats['GP'][0]  # Games played

ast_per_game = stats['AST'][0] / gp
reb_per_game = stats['REB'][0] / gp

print(f"AST per game: {ast_per_game:.1f}")
print(f"REB per game: {reb_per_game:.1f}")
```
